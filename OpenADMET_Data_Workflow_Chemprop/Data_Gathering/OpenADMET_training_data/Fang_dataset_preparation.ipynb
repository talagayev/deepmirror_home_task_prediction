```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("raw_datasets/fang/Fang_dataset_3521_compounds.csv")
```

```python
df
```

```python
df_remove = df[["SMILES", "LOG HLM_CLint (mL/min/kg)"]].dropna().reset_index(drop=True)
```

```python
df_remove
```

```python
df_remove["hlm_value_unscaled"] = 10 ** df_remove["LOG HLM_CLint (mL/min/kg)"]
df_remove["hlm_value_unscaled_log10"] = df_remove["LOG HLM_CLint (mL/min/kg)"]

df_remove["hlm_value_scaled"] = 10 ** df_remove["LOG HLM_CLint (mL/min/kg)"]
df_remove["hlm_value_scaled_log10_high"] = df_remove["LOG HLM_CLint (mL/min/kg)"]
df_remove["hlm_value_scaled_log10_low"] = df_remove["LOG HLM_CLint (mL/min/kg)"]
```

```python
df_remove
```

## Plot Data to see distribution

```python
import matplotlib.pyplot as plt

# Mean values
mean_log1p = df_remove["hlm_value_scaled_log10_high"].mean()
mean_value = df_remove["hlm_value_scaled_log10_low"].mean()

# Plot 1: distribution of hlm_value_log10
plt.figure(figsize=(8, 5))
plt.hist(df_remove["hlm_value_scaled_log10_high"], bins=30)
plt.axvline(mean_log1p, linestyle="--", label=f"Mean = {mean_log1p:.3f}")
plt.xlabel("HLM CLint log10")
plt.ylabel("number of molecules")
plt.title("Distribution of Fang Dataset log10 high")
plt.legend()
plt.show()

# Plot 2: distribution of hlm_value
plt.figure(figsize=(8, 5))
plt.hist(df_remove["hlm_value_scaled_log10_low"], bins=30)
plt.axvline(mean_value, linestyle="--", label=f"Mean = {mean_value:.3f}")
plt.xlabel("HLM CLint log10")
plt.ylabel("number of molecules")
plt.title("Distribution of Fang Dataset log10 low")
plt.legend()
plt.show()
```

```python
df_remove["dataset"] = "fang_dataset"
df_remove["data_year"] = "2023"
df_remove.to_csv("adjusted_datasets/fang_dataset_adjusted_new.csv")
```

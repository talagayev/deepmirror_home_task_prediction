```python
import pandas as pd
```

```python
df_fang = pd.read_csv("adjusted_datasets/fang_dataset_adjusted_new.csv")
```

```python
df_fang
```

```python
df_fang["mol_id"] = [f"fang_{i}" for i in range(1, len(df_fang) + 1)]
df_fang["scaling_was_applied"] = "False" 
df_fang
```

```python
df_openadmet_expansion = pd.read_csv("adjusted_datasets/openadmet_expansion_train_adjusted_new.csv")
```

```python
df_openadmet_expansion
```

```python
df_openadmet_expansion = df_openadmet_expansion.rename(columns={"Molecule Name": "mol_id"})
df_openadmet_expansion["scaling_was_applied"] = "False" 
df_openadmet_expansion
```

```python
df_openadmet_polaris = pd.read_csv("adjusted_datasets/antiviral_admet_2025_unblinded_adjusted_new.csv")
```

```python
df_openadmet_polaris
```

```python
df_openadmet_polaris = df_openadmet_polaris.rename(columns={"CXSMILES": "SMILES"})
df_openadmet_polaris = df_openadmet_polaris.rename(columns={"Molecule Name": "mol_id"})
df_openadmet_polaris["scaling_was_applied"] = "True" 
df_openadmet_polaris
```

```python
df_chembl = pd.read_csv("adjusted_datasets/chembl_hlm_adjusted_new.csv")
```

```python
df_chembl
```

```python
df_chembl = df_chembl.rename(columns={"smiles": "SMILES"})
df_chembl = df_chembl.rename(columns={"molecule_chembl_id": "mol_id"})
df_chembl["scaling_was_applied"] = df_chembl["standard_units"].map({
    "mL.min-1.kg-1": False,
    "mL.min-1.g-1": True
})
df_chembl
```

```python
combined_df = pd.concat([df_fang, df_openadmet_polaris, df_openadmet_expansion, df_chembl], ignore_index=True)
```

```python
combined_df
```

```python
combined_df_filtered = combined_df[["mol_id", "SMILES", "hlm_value_unscaled", "hlm_value_unscaled_log10", "hlm_value_scaled", "hlm_value_scaled_log10_low", "hlm_value_scaled_log10_high", "dataset", "scaling_was_applied"]]
```

```python
combined_df_filtered
```

```python
(combined_df_filtered["hlm_value_unscaled"] > 1000).sum()
```

```python
import matplotlib.pyplot as plt

# Mean values
mean_scaled = combined_df_filtered["hlm_value_scaled_log10_high"].mean()
mean_scaled_log1p = combined_df_filtered["hlm_value_scaled_log10_low"].mean()

# Plot 1: scaled
plt.figure(figsize=(8, 5))
plt.hist(combined_df_filtered["hlm_value_scaled_log10_high"], bins=30)
plt.axvline(mean_scaled, linestyle="--", label=f"Mean = {mean_scaled:.3f}")
plt.xlabel("HLM CLint Scaled log10")
plt.ylabel("number of molecules")
plt.title("Distribution of HLM Scaled log10 high")
plt.legend()
plt.show()

# Plot 2: scaled log10 nan
plt.figure(figsize=(8, 5))
plt.hist(combined_df_filtered["hlm_value_scaled_log10_low"], bins=30)
plt.axvline(mean_scaled_log1p, linestyle="--", label=f"Mean = {mean_scaled_log1p:.3f}")
plt.xlabel("HLM CLint Scaled log10")
plt.ylabel("number of molecules")
plt.title("Distribution of HLM Scaled log10 low")
plt.legend()
plt.show()
```

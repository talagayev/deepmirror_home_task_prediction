The ChEMBL retrieved data contains values with various units. While we retain the applicable mL.min-1.kg-1 and mL.min-1.g-1 which we need to scale we drop the remainder, including ml/min.mg due to it only having 5 data points and min.kg/L having 4 data points due to them requiring a separate scaling algorithm most likely

```python
df_inside_range["standard_units"].unique()
```

```python
df_inside_range["standard_units"].value_counts().reindex([
    'mL.min-1.kg-1', 'mL.min-1.g-1', 'nmol/mg/min', 'nM/min/mg',
    'ml/min.mg', 'mg/ml/g', 'min.kg/L', 'hr.Kg/L', 'uL/ml/kg',
    'uL/ml.millig', 'ug ml-1', 'ug/min/mg', 'uL/min', 'ug/min',
    'L/min'
], fill_value=0)
```

```python
df_inside_range = df_inside_range[
    df_inside_range["standard_units"].isin([
        "mL.min-1.kg-1",
        "mL.min-1.g-1",
    ])
].reset_index(drop=True)
```

```python
df_inside_range
```

We also need to drop the 0 value data points due to them influencing log10 and with the OpenADMET not having 0 cases

```python
df_inside_range
```

```python
df_inside_range = df_inside_range[df_inside_range["document_chembl_id"] != "CHEMBL5442168"]
```

```python
def clip_and_log_transform(y: np.ndarray):
    """
    Clip to a detection limit and transform to log10 scale.
    Parameters
    ----------
    y : np.ndarray
        The array to be clipped and transformed.
    """
    y = np.clip(y, a_min=0, a_max=None)
    return np.log10(y + 1)
```

```python
df_inside_range
```

```python
import numpy as np

df_inside_range["hlm_value_unscaled"] = df_inside_range["standard_value"]

df_inside_range["hlm_value_scaled_high"] = np.where(
    df_inside_range["standard_units"] == "mL.min-1.g-1",
    df_inside_range["standard_value"] * 1.028,
    df_inside_range["standard_value"]
)

df_inside_range["hlm_value_scaled_low"] = np.where(
    df_inside_range["standard_units"] == "mL.min-1.g-1",
    df_inside_range["standard_value"] * 0.840,
    df_inside_range["standard_value"]
)
```

```python
df_inside_range
```

```python
df_inside_range["hlm_value_unscaled_log10"] = clip_and_log_transform(
    df_inside_range["hlm_value_unscaled"].to_numpy()
)

df_inside_range["hlm_value_scaled_log10_high"] = clip_and_log_transform(
    df_inside_range["hlm_value_scaled_high"].to_numpy()
)

df_inside_range["hlm_value_scaled_log10_low"] = clip_and_log_transform(
    df_inside_range["hlm_value_scaled_low"].to_numpy()
)
```

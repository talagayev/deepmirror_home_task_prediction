```python
df_remove = df[["Molecule Name", "SMILES", "HLM CLint"]].dropna().reset_index(drop=True)
```

Some molecules have 0, which will make log10 not work with them, will give an empty value. possible to either add a small value to not lose the value in log10 or use maybe log. here we will clip it to the closest non zero value

```python
df_remove["hlm_value_unscaled"] = df_remove["HLM CLint"]
```

```python
def clip_and_log_transform(y: np.ndarray):
    """
    Clip to a detection limit and transform to log10 scale.
    Parameters
    ----------
    y : np.ndarray
        The array to be clipped and transformed.
    """
    y = np.clip(y, a_min=0, a_max=None)
    return np.log10(y + 1)
```

```python
df_remove["hlm_value_unscaled_log10"] = clip_and_log_transform(
    df_remove["hlm_value_unscaled"].to_numpy()
)
```

```python
df_remove["hlm_value_scaled_log10_high"] = df_remove["hlm_value_unscaled_log10"]
df_remove["hlm_value_scaled_log10_low"] = df_remove["hlm_value_unscaled_log10"]
```

```python
import pandas as pd
```

```python
df_fang = pd.read_csv("adjusted_datasets/fang_dataset_adjusted.csv")
```

```python
df_fang
```

```python
df_fang["mol_id"] = [f"fang_{i}" for i in range(1, len(df_fang) + 1)]
df_fang["scaling_was_applied"] = "False" 
df_fang
```

```python
df_openadmet_expansion = pd.read_csv("adjusted_datasets/openadmet_expansion_train_adjusted.csv")
```

```python
df_openadmet_expansion
```

```python
df_openadmet_expansion = df_openadmet_expansion.rename(columns={"Molecule Name": "mol_id"})
df_openadmet_expansion["scaling_was_applied"] = "False" 
df_openadmet_expansion
```

```python
df_openadmet_polaris = pd.read_csv("adjusted_datasets/antiviral_admet_2025_unblinded_adjusted.csv")
```

```python
df_openadmet_polaris
```

```python
df_openadmet_polaris = df_openadmet_polaris.rename(columns={"CXSMILES": "SMILES"})
df_openadmet_polaris = df_openadmet_polaris.rename(columns={"Molecule Name": "mol_id"})
df_openadmet_polaris["scaling_was_applied"] = "True" 
df_openadmet_polaris
```

```python
df_chembl = pd.read_csv("adjusted_datasets/chembl_hlm_adjusted.csv")
```

```python
df_chembl
```

```python
df_chembl = df_chembl.rename(columns={"smiles": "SMILES"})
df_chembl = df_chembl.rename(columns={"molecule_chembl_id": "mol_id"})
df_chembl["scaling_was_applied"] = df_chembl["standard_units"].map({
    "mL.min-1.kg-1": False,
    "mL.min-1.g-1": True
})
df_chembl
```

```python
combined_df = pd.concat([df_fang, df_openadmet_polaris, df_openadmet_expansion, df_chembl], ignore_index=True)
```

```python
combined_df
```

```python
combined_df_filtered = combined_df[["mol_id", "SMILES", "hlm_value_unscaled", "hlm_value_unscaled_log1p", "hlm_value_scaled", "hlm_value_scaled_log1p", "dataset", "scaling_was_applied"]]
```

```python
combined_df_filtered
```

```python
combined_df_filtered.to_csv("combined_datasets/combined_dataset_training.csv")
```

```python
mask = (
    combined_df_filtered["scaling_was_applied"]
    .astype(str)
    .str.strip()
    .str.lower()
    .eq("false")
)

combined_df_filtered_safe_data = combined_df_filtered.loc[mask]
```

```python
combined_df_filtered_safe_data
```

```python
combined_df_filtered_safe_data["dataset"].unique()
```

```python
combined_df_filtered_safe_data.to_csv("combined_datasets/combined_dataset_training_no_scaling.csv")
```

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("raw_datasets/polaris/antiviral_admet_2025_unblinded.csv")
```

```python
df
```

Let's first drop the rows that don't have HLM values

```python
df_remove = df[["CXSMILES", "HLM", "Molecule Name", "Set"]].dropna().reset_index(drop=True)
```

```python
df_remove
```

On the Polaris Dataset page the organizers wrote that for 4 compounds for HLM, the values were not correct since they changed after time, but they kept them for benchmarking. Due to them being incorrect and kept for benchmarking we will remove them from here

```python
to_remove = [
    "CNC(=O)CN1C[C@@]2(C(=O)N(C3=CN=CC4=CC=CC=C34)C[C@@H]2N)C2=CC(Cl)=CC=C2C1=O |a:7,22|",
    "CNCCN1C[C@@]2(C(=O)N(C3=CN=CC4=CC=CC=C34)C[C@@H]2CS(=O)(=O)N(C)C)C2=CC(Cl)=CC=C2C1=O |a:6,21|",
    "C[C@H]1CN(C2=CN=CC3=CC=CC=C23)C(=O)[C@@]12CN(CC1=NC=CO1)C(=O)C1=CC=C(F)C=C12 |a:1,16|",
    "C[C@H]1CN(C2=CN=CC3=CC=CC=C23)C(=O)[C@@]12CN(CC1=NN(C)C=N1)C(=O)C1=CC=C(F)C=C12 |a:1,16|",
]

df_remove = df_remove[
    ~df_remove["CXSMILES"].isin(to_remove) &
    df_remove["HLM"].notna()
].reset_index(drop=True)
```

```python
df_remove
```

Let's add the scaled and unscaled values in their respective columns to be compatible with the rest of the molecules. also drop the two datapoints with 0

```python
df_remove["hlm_value_unscaled"] = df_remove["HLM"]

# true log10, zeros become NaN
df_remove["hlm_value_unscaled_log10_nan"] = np.where(
    df_remove["hlm_value_unscaled"] > 0,
    np.log10(df_remove["hlm_value_unscaled"]),
    np.nan
)

# epsilon version, zeros kept via small offset
epsilon = 1e-1
df_remove["hlm_value_unscaled_log10_epsilon"] = np.where(
    df_remove["hlm_value_unscaled"] == 0,
    np.log10(epsilon),
    np.log10(df_remove["hlm_value_unscaled"])
)
```

```python
df_remove["hlm_value_scaled"] = df_remove["HLM"] * 1.028

# true log10, zeros become NaN
df_remove["hlm_value_scaled_log10_nan"] = np.where(
    df_remove["hlm_value_scaled"] > 0,
    np.log10(df_remove["hlm_value_scaled"]),
    np.nan
)

# epsilon version, zeros kept via small offset
epsilon = 1e-1
df_remove["hlm_value_scaled_log10_epsilon"] = np.where(
    df_remove["hlm_value_scaled"] == 0,
    np.log10(epsilon),
    np.log10(df_remove["hlm_value_scaled"])
)
```

```python
df_remove
```

```python
import matplotlib.pyplot as plt

# Mean values
mean_scaled = df_remove["hlm_value_scaled"].mean()
mean_scaled_log10_nan = df_remove["hlm_value_scaled_log10_nan"].mean()
mean_scaled_log10_epsilon = df_remove["hlm_value_scaled_log10_epsilon"].mean()

# Plot 1: scaled
plt.figure(figsize=(8, 5))
plt.hist(df_remove["hlm_value_scaled"], bins=30)
plt.axvline(mean_scaled, linestyle="--", label=f"Mean = {mean_scaled:.3f}")
plt.xlabel("HLM CLint Scaled")
plt.ylabel("number of molecules")
plt.title("Distribution of HLM Scaled")
plt.legend()
plt.show()

# Plot 2: scaled log10 nan
plt.figure(figsize=(8, 5))
plt.hist(df_remove["hlm_value_unscaled"], bins=30)
plt.axvline(mean_scaled_log10_nan, linestyle="--", label=f"Mean = {mean_scaled_log10_nan:.3f}")
plt.xlabel("HLM CLint Scaled log10 Nan")
plt.ylabel("number of molecules")
plt.title("Distribution of HLM Scaled log10")
plt.legend()
plt.show()

# Plot 3: scaled log10 epsilon
plt.figure(figsize=(8, 5))
plt.hist(df_remove["hlm_value_scaled_log10_epsilon"], bins=30)
plt.axvline(mean_scaled_log10_epsilon, linestyle="--", label=f"Mean = {mean_scaled_log10_epsilon:.3f}")
plt.xlabel("HLM CLint Scaled Log10 Epsilon")
plt.ylabel("number of molecules")
plt.title("Distribution of HLM Scaled Log10")
plt.legend()
plt.show()
```

```python
df_remove["dataset"] = "openadmet_polaris"
df_remove["data_year"] = "2025"
df_remove.to_csv("adjusted_datasets/antiviral_admet_2025_unblinded_adjusted.csv")
```
